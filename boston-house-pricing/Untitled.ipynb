```python
# independent and depenedent features

x = df.iloc[:,:-1]
y = df.iloc[:,-1]
```

```python
df.columns
```

```python
x.head()
```

```python
y.head()
```

```python
#train_test split

from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.33, random_state=42)
```

```python
x_test.shape
```

```python
x_train.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
# statnderdise the dataset
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
```

```python
x_train=scaler.fit_transform(x_train)
```

```python
# to avoid data leakage we only transform the X-test 
x_test = scaler.transform(x_test)
```

```python
x_test
```

## model training

```python
from sklearn.linear_model import LinearRegression
```

```python
liner = LinearRegression()
```

```python
liner
```

```python
liner.fit(x_train, y_train)
```

```python
# print the coffesients and intercept
```

```python
print(liner.coef_)
```

```python
print(liner.intercept_)
```

```python
liner_pred = liner.predict(x_test)
```

```python
liner_pred
```

## asssumptions of linear regression

```python
sns.scatterplot(x=y_test,y=liner_pred)
```

```python
residual = y_test - liner_pred
```

```python
residual
```

```python
sns.kdeplot(residual)
```

```python
## scatter plot with preiction and residual
# uniform distribution
# homocidacity
sns.scatterplot(x=liner_pred, y=residual)
```

```python
# performance metrics
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

```python
print(mean_squared_error(y_test, liner_pred))
```

```python
print(mean_absolute_error(y_test, liner_pred))
```

```python
print(np.sqrt(mean_squared_error(y_test, liner_pred)))
```

```python
# r square and adj r sqr

from sklearn.metrics import r2_score
```

```python
r2_score(y_test, liner_pred)
```

```python
# ridge regression

from sklearn.linear_model import Ridge
ridge = Ridge()
```

```python
ridge.fit(x_train,y_train)
```

```python
ridge_pred = ridge.predict(x_test)
```

```python
r2_score(y_test, ridge_pred)
```
